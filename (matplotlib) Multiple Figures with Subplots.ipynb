# Creating Subplots in matplotlib

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10,10)
```

```python
x1 = range(10)
y1 = range(10)
```

# Create a figure with subplots

```python
fig1, ax1 = plt.subplots(nrows=2, ncols=2)
```

## Plot in one of the subplots

```python
ax1[0, 1].plot(x1, y1)
```

```python
fig1
```

# Create another figure with subplots

```python
x2 = range(10, 20, 1)
y2 = range(10, 20, 1)
```

```python
fig2, ax2 = plt.subplots(nrows=2, ncols=2)
```

## Plot in one of the subplots

```python
ax2[0,0].plot(x2, y2)
```

```python
fig2
```

### If you want to clear a subplot and plot new data in the first figure

```python
ax1[0,1].cla()
```

```python
fig1
```
